## **Log Transform**

**When to Apply Log Transform:**

1.Skewed Data: To reduce right-skewness and make data more normal.

2.Exponential Growth: To linearize relationships with exponential trends.

3.Heteroscedasticity: To stabilize variance in regression analysis.
Outliers: To reduce the impact of extreme values.

4.Wide Range Data: To handle variables spanning several orders of magnitude.

```python
# log Transform
df['Age_log'] = np.log(df['Age'])
```

```python
sns.distplot(df['Age_log'])
plt.show()
```

```python
df['Age_log'].skew()
```

```python
# apply qq-plot
sm.qqplot(df['Age_log'], line='s')
plt.show()
```

## **Square Transform**

**When to Use Square Transformation**

1.Left-Skewed Data: To reduce left skewness and make data more symmetric.

2.Enhance Differences: Amplifies differences between smaller values.

3.Linear Relationships: To model relationships where the dependent variable grows quadratically with the independent variable.

4.Variance Stabilization: To reduce variability in some datasets.
Increase Spread: Useful when smaller values need more emphasis.

Note: Be cautious with large values, as squaring can exaggerate their impact.

## **Square Root Transform**

**When to Use Square Root Transformation **
1.Right-Skewed Data: To reduce moderate right skewness.

2.Stabilize Variance: Helps reduce heteroscedasticity in data.

3.Count Data: Works well with count variables (e.g., number of events).
Dampen Large Values: Reduces the impact of larger values without compressing them too much.

4.Linearize Growth: Suitable when the relationship follows a square root pattern.

Note: Data should be non-negative; if zeros are present, add a small constant before transformation.

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.preprocessing import FunctionTransformer
trf = FunctionTransformer(func=np.log1p)

X_train_trf = trf.fit_transform(X_train)
X_test_trf = trf.transform(X_test)
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf1.fit(X_train_trf,y_train)
clf2.fit(X_train_trf,y_train)

print()
```

```python
from sklearn.metrics import accuracy_score,corss

print('Logistic Regression Accuracy:',accuracy_score(y_test,clf1.predict(X_test_trf)))
print('Decision Tree Accuracy:',accuracy_score(y_test,clf2.predict(X_test_trf)))
```

```python
X_tranformed = trf.fit_transform(df.drop('Survived',axis=1))

clf = LogisticRegression()
clf2 = DecisionTreeClassifier()
clf.fit(X_tranformed,df['Survived'])

```
